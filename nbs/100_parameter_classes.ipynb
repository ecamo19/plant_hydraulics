```python
# | export
@dataclass
class PhysCon:
    """Object holding Physical constants

    - grav:  Gravitational acceleration (m/s2)
    - tfrz:  Freezing point of water (K)
    - sigma: Stefan-Boltzmann constant (W/m2/K4)
    - mmdry: Molecular mass of dry air (kg/mol)
    - mmh2o: Molecular mass of water (kg/mol)
    - cpd:   Specific heat of dry air at constant pressure (J/kg/K)
    - cpw:   Specific heat of water vapor at constant pressure (J/kg/K)
    - rgas:  Universal gas constant (J/K/mol)
    - visc0: Kinematic viscosity at 0C and 1013.25 hPa (m2/s)
    - Dh0:   Molecular diffusivity (heat) at 0C and 1013.25 hPa (m2/s)
    - Dv0:   Molecular diffusivity (H2O) at 0C and 1013.25 hPa (m2/s)
    - Dc0:   Molecular diffusivity (CO2) at 0C and 1013.25 hPa (m2/s)
    - denh2o: Density of liquid water (kg/m3)
    """

    # Gravitational acceleration (m/s2)
    grav: float = 9.80665

    # Freezing point of water (K)
    tfrz: float = 273.15

    # Stefan-Boltzmann constant (W/m2/K4)
    sigma: float = 5.67e-08

    # Molecular mass of dry air (kg/mol)
    mmdry: float = 28.97e-03

    # Molecular mass of water (kg/mol)
    mmh2o: float = 18.02e-03

    # Specific heat of dry air at constant pressure (J/kg/K)
    cpd: float = 1005.0

    # Specific heat of water vapor at constant pressure (J/kg/K)
    cpw: float = 1846.0

    # Universal gas constant (J/K/mol)
    rgas: float = 8.31446

    # Kinematic viscosity at 0C and 1013.25 hPa (m2/s)
    visc0: float = 13.3e-06

    # Molecular diffusivity (heat) at 0C and 1013.25 hPa (m2/s)
    Dh0: float = 18.9e-06

    # Molecular diffusivity (H2O) at 0C and 1013.25 hPa (m2/s)
    Dv0: float = 21.8e-06

    # Molecular diffusivity (CO2) at 0C and 1013.25 hPa (m2/s)
    Dc0: float = 13.8e-06

    # Density of liquid water (kg/m3)
    denh2o: float = 1000.0
```

```python
# | export


@dataclass
class Leaf:
    """Object holding Leaf physiological and structural parameters

    - c3psn: Photosynthetic pathway: 1 = C3, 0 = C4
    - colim: Photosynthesis co-limitation: 0 = no, 1 = yes
    - vcmax25: Maximum carboxylation rate at 25C (umol/m2/s)
    - jmax25: Maximum electron transport rate at 25C (umol/m2/s)
    - rd25: Leaf respiration rate at 25C (umol CO2/m2/s)
    - kc25: Michaelis-Menten constant for CO2 at 25C (umol/mol)
    - ko25: Michaelis-Menten constant for O2 at 25C (mmol/mol)
    - cp25: CO2 compensation point at 25C (umol/mol)
    - kcha: Activation energy for Kc (J/mol)
    - koha: Activation energy for Ko (J/mol)
    - cpha: Activation energy for Cp (J/mol)
    - vcmaxha: Activation energy for Vcmax (J/mol)
    - jmaxha: Activation energy for Jmax (J/mol)
    - rdha:  Activation energy for Rd (J/mol)
    - vcmaxhd: Deactivation energy for Vcmax (J/mol)
    - jmaxhd:  Deactivation energy for Jmax (J/mol)
    - rdhd:  Deactivation energy for Rd (J/mol)
    - vcmaxse: Entropy term for Vcmax (J/mol/K)
    - jmaxse: Entropy term for Jmax (J/mol/K)
    - rdse: Entropy term for Rd (J/mol/K)
    - vcmaxc: Vcmax scaling factor for high temperature inhibition
    - jmaxc:  Jmax scaling factor for high temperature inhibition
    - rdc:  Rd scaling factor for high temperature inhibition
    - phi_psii: Quantum yield of PS II
    - theta_j: Empirical curvature parameter for electron transport rate
    - colim_c3: Empirical curvature parameter for C3 co-limitation
    - colim_c4a:  Empirical curvature parameter for C4 co-limitation
    - colim_c4b:  Empirical curvature parameter for C4 co-limitation
    - qe_c4: C4: Quantum yield (mol CO2 / mol photons)
    - kp25_c4: C4: Initial slope of CO2 response curve at 25C (mol/m2/s)
    - dleaf: Leaf dimension (m)
    - emiss: Leaf emissivity
    - rho: Leaf reflectance [vis, nir]
    - tau: Leaf transmittance [vis, nir]
    - iota: Stomatal efficiency (umol CO2/ mol H2O)
    - capac: Plant capacitance (mmol H2O/m2 leaf area/MPa)
    - minl_wp: Minimum leaf water potential (MPa)
    - gplant: Stem (xylem-to-leaf) hydraulic conductance (mmol H2O/m2 leaf area/s/MPa)
    - g0: Minimum Conductance (mol/m2/s)
    - g1_medlyn: Medlyn slope parameter (kPa^0.5). Controls stomatal sensitivity to VPD. Higher g1 = less conservative.
    - stomatal_model: Stomatal model to use "optimization", "medlyn" or "profit_max"
    - a_psi: Parameter that controls the steepness of this curve in the "profit_max" model
    - psi_50: Water potential at 50% loss of conductance (MPa). Always < 0.
    """

    # Photosynthetic pathway: 1 = C3, 0 = C4
    c3psn: int = 1

    # Photosynthesis co-limitation: 0 = no, 1 = yes
    colim: int = 1

    # Maximum carboxylation rate at 25C (umol/m2/s)
    vcmax25: float = 0.0

    # Maximum electron transport rate at 25C (umol/m2/s)
    jmax25: float = 0.0

    # Leaf respiration rate at 25C (umol CO2/m2/s)
    rd25: float = 0.0

    # Michaelis-Menten constant for CO2 at 25C (umol/mol)
    kc25: float = 0.0

    # Michaelis-Menten constant for O2 at 25C (mmol/mol)
    ko25: float = 0.0

    # CO2 compensation point at 25C (umol/mol)
    cp25: float = 0.0

    # Activation energy for Kc (J/mol)
    kcha: float = 0.0

    # Activation energy for Ko (J/mol)
    koha: float = 0.0

    # Activation energy for Cp (J/mol)
    cpha: float = 0.0

    # Activation energy for Vcmax (J/mol)
    vcmaxha: float = 0.0

    # Activation energy for Jmax (J/mol)
    jmaxha: float = 0.0

    # Activation energy for Rd (J/mol)
    rdha: float = 0.0

    # Deactivation energy for Vcmax (J/mol)
    vcmaxhd: float = 0.0

    # Deactivation energy for Jmax (J/mol)
    jmaxhd: float = 0.0

    # Deactivation energy for Rd (J/mol)
    rdhd: float = 0.0

    # Entropy term for Vcmax (J/mol/K)
    vcmaxse: float = 0.0

    # Entropy term for Jmax (J/mol/K)
    jmaxse: float = 0.0

    # Entropy term for Rd (J/mol/K)
    rdse: float = 0.0

    # Vcmax scaling factor for high temperature inhibition
    vcmaxc: float = 0.0

    # Jmax scaling factor for high temperature inhibition
    jmaxc: float = 0.0

    # Rd scaling factor for high temperature inhibition
    rdc: float = 0.0

    # Quantum yield of PS II
    phi_psii: float = 0.0

    # Empirical curvature parameter for electron transport rate
    theta_j: float = 0.0

    # Empirical curvature parameter for C3 co-limitation
    colim_c3: float = 0.0

    # Empirical curvature parameter for C4 co-limitation
    colim_c4a: float = 0.0

    # Empirical curvature parameter for C4 co-limitation
    colim_c4b: float = 0.0

    # C4: Quantum yield (mol CO2 / mol photons)
    qe_c4: float = 0.0

    # C4: Initial slope of CO2 response curve at 25C (mol/m2/s)
    kp25_c4: float = 0.0

    # Leaf dimension (m)
    dleaf: float = 0.0

    # Leaf emissivity
    emiss: float = 0.0

    # Leaf reflectance [vis, nir]
    rho: list = field(default_factory=lambda: [0.0, 0.0])

    # Leaf transmittance [vis, nir]
    tau: list = field(default_factory=lambda: [0.0, 0.0])

    # Stomatal efficiency (umol CO2/ mol H2O)
    iota: float = 0.0

    # Plant capacitance (mmol H2O/m2 leaf area/MPa)
    capac: float = 0.0

    # Minimum leaf water potential (MPa)
    minl_wp: float = 0.0

    # Stem (xylem-to-leaf) hydraulic conductance (mmol H2O/m2 leaf area/s/MPa)
    gplant: float = 0.0

    # Which stomatal model to use:
    #   "optimization" = original WUE optimization from Bonan Ch. 13
    #                    (uses brent_root + stomata_efficiency)
    #   "medlyn"       = Medlyn et al. (2011) empirical model
    #                    (uses iterative gs-An coupling + hard hydraulic cap)
    #   "profit_max"   = Medlyn + Sperry profit-maximisation framework
    #                    (uses gs sweep + gain-cost optimisation)
    stomatal_model: Literal["optimization", "medlyn", "profit_max"] = "optimization"

    # Minimum stomatal conductance (mol H2O/m2/s).
    # The residual conductance when An = 0 (e.g., at night).
    # Example values:
    #   Broadleaf trees: 0.01 mol/m2/s
    #   Conifers:        0.005 mol/m2/s
    #   Crops:           0.02 mol/m2/s
    g0: float = 0.01

    # Medlyn slope parameter (kPa^0.5).
    # Controls stomatal sensitivity to VPD. Higher g1 = less conservative.
    # Example values from:
    #   Tropical broadleaf evergreen: 3.77 kPa^0.5
    #   Temperate broadleaf deciduous: 4.45 kPa^0.5
    #   Needleleaf evergreen:          2.35 kPa^0.5
    #   C3 crops/grasses:              5.79 kPa^0.5
    #   C4 crops/grasses:              1.62 kPa^0.5
    g1_medlyn: float = 4.0

    # PROFIT-MAX: Vulnerability curve shape parameter
    #
    # WHY: The profit-maximisation model needs a vulnerability curve that
    # describes how hydraulic conductance declines with water potential.
    # The a_psi parameter controls the steepness of this curve.

    # Low a_psi (2-3): Represents a gentle curve, therefore conductance loss
    # begins early and spreads over a wide psi range. Represents species that
    # close stomata cautiously to avoid any damage.

    # High a_psi (5-8): Represents a steep curve, therefore conductance is
    # maintained until psi is very close to minl_wp, then drops sharply.
    # Represents a species that tolerate more stress.

    # Default 4: A moderate steepness suitable for a "typical" broadleaf tree.
    #
    # Note: Only used when stomatal_model = "profit_max".
    #   Ignored by "optimization" and "medlyn" models.
    a_psi: float = 4.0

    # PROFIT-MAX: Water potential at 50% loss of hydraulic conductance (MPa).
    #
    # This is the TRUE P50 of the vulnerability curve — a separate plant trait
    # from minl_wp (the critical/minimum safe psi).
    #
    # Distinction:
    #   psi_50  = where 50% of conductance is lost. Controls the SHAPE of the
    #             cost function. Species with more negative psi_50 tolerate
    #             drier conditions before losing conductance.
    #   minl_wp = the hard safety threshold. Beyond this, cavitation is lethal.
    #             Typically near P88–P95.
    #
    # Typical values (temperate broadleaf):
    #   psi_50  ≈ -2.5 MPa  (moderate vulnerability, e.g. Quercus robur)
    #   minl_wp ≈ -3.0 MPa  (safety margin: ~0.5 MPa above psi_50)
    #
    # Rule of thumb: minl_wp should be 1.5–2× more negative than psi_50.
    # If psi_50 = minl_wp, cost at the safety threshold is always 0.5 by
    # construction — the model becomes too permissive.
    #
    # Note: Only used when stomatal_model = "profit_max".
    psi_50: float = -2.5
```

```python
# | export


@dataclass
class Flux:
    """Object holding Flux variables (state and output)

    - __Site / canopy__

        - height: Leaf height (m)
        - lai: Canopy leaf area index (m2/m2)

    - __Hydraulics__

        - rplant: Aboveground plant stem resistance (MPa.s.m2/mmol H2O)
        - rsoil: Soil hydraulic resistance (MPa.s.m2/mmol H2O)
        - lsc: Leaf-specific conductance (mmol H2O/m2 leaf/s/MPa)
        - psi_soil: Weighted soil water potential (MPa)
        - psi_leaf: Leaf water potential (MPa)
        - et_loss: Fraction of transpiration from each soil layer

    - __Radiation__

        - swinc: Incident solar radiation on leaf [vis, nir]
        - swflx: Absorbed solar radiation [vis, nir]
        - apar: Absorbed PAR (umol photon/m2 leaf/s)
        - qa: Leaf radiative forcing (W/m2 leaf)

    - __Leaf temperature and energy balance__

        - tleaf: Leaf temperature (K)
        - rnet: Leaf net radiation (W/m2 leaf)
        - lwrad: Longwave radiation emitted from leaf (W/m2 leaf)
        - shflx: Leaf sensible heat flux (W/m2 leaf)
        - lhflx: Leaf latent heat flux (W/m2 leaf)
        - etflx: Leaf transpiration flux (mol H2O/m2 leaf/s)

    - __Boundary layer conductances__

        - gbh: Boundary layer conductance, heat (mol/m2 leaf/s)
        - gbv: Boundary layer conductance, H2O (mol H2O/m2 leaf/s)
        - gbc: Boundary layer conductance, CO2 (mol CO2/m2 leaf/s)

    - __Stomatal conductance__

        - gs: Leaf stomatal conductance (mol H2O/m2 leaf/s)

    - __Photosynthesis__

        - vcmax: Maximum carboxylation rate (umol/m2/s)
        - jmax: Maximum electron transport rate (umol/m2/s)
        - cp: CO2 compensation point (umol/mol)
        - kc: Michaelis-Menten constant for CO2 (umol/mol)
        - ko: Michaelis-Menten constant for O2 (mmol/mol)
        - je: Electron transport rate (umol/m2/s)
        - kp_c4: C4: Initial slope of CO2 response curve (mol/m2/s)
        - ac: Rubisco-limited gross photosynthesis (umol CO2/m2 leaf/s)
        - rd: Leaf respiration rate (umol CO2/m2 leaf/s)
        - aj: RuBP regeneration-limited gross photosynthesis (umol CO2/m2 leaf/s)
        - ap: Product-limited (C3) or CO2-limited (C4) gross photosynthesis
        - ag: Leaf gross photosynthesis (umol CO2/m2 leaf/s)
        - an: Leaf net photosynthesis (umol CO2/m2 leaf/s)
        - cs: Leaf surface CO2 (umol/mol)
        - ci: Leaf intercellular CO2 (umol/mol)
        - hs: Leaf fractional humidity at surface (dimensionless)
        - vpd: Leaf vapor pressure deficit at surface (Pa)

    - __Time step__

        - dt: Model time step (s)
    """

    # Site / canopy -------------------------------------------------------------

    # Leaf height (m)
    height: float = 0.0

    # Canopy leaf area index (m2/m2)
    lai: float = 0.0

    # Hydraulics ----------------------------------------------------------------

    # Aboveground plant stem resistance (MPa.s.m2/mmol H2O)
    rplant: float = 0.0

    # Soil hydraulic resistance (MPa.s.m2/mmol H2O)
    rsoil: float = 0.0

    # Leaf-specific conductance (mmol H2O/m2 leaf/s/MPa)
    lsc: float = 0.0

    # Weighted soil water potential (MPa)
    psi_soil: float = 0.0

    # Leaf water potential (MPa)
    psi_leaf: float = 0.0

    # Fraction of transpiration from each soil layer
    et_loss: list = field(default_factory=list)

    # Radiation -----------------------------------------------------------------

    # Incident solar radiation on leaf [vis, nir]
    swinc: list = field(default_factory=lambda: [0.0, 0.0])

    # Absorbed solar radiation [vis, nir]
    swflx: list = field(default_factory=lambda: [0.0, 0.0])

    # Absorbed PAR (umol photon/m2 leaf/s)
    apar: float = 0.0

    # Leaf radiative forcing (W/m2 leaf)
    qa: float = 0.0

    # Leaf temperature and energy balance ---------------------------------------

    # Leaf temperature (K)
    tleaf: float = 0.0

    # Leaf net radiation (W/m2 leaf)
    rnet: float = 0.0

    # Longwave radiation emitted from leaf (W/m2 leaf)
    lwrad: float = 0.0

    # Leaf sensible heat flux (W/m2 leaf)
    shflx: float = 0.0

    # Leaf latent heat flux (W/m2 leaf)
    lhflx: float = 0.0

    # Leaf transpiration flux (mol H2O/m2 leaf/s)
    etflx: float = 0.0

    # Boundary layer conductances -----------------------------------------------

    # Boundary layer conductance, heat (mol/m2 leaf/s)
    gbh: float = 0.0

    # Boundary layer conductance, H2O (mol H2O/m2 leaf/s)
    gbv: float = 0.0

    # Boundary layer conductance, CO2 (mol CO2/m2 leaf/s)
    gbc: float = 0.0

    # Stomatal conductance ------------------------------------------------------

    # Leaf stomatal conductance (mol H2O/m2 leaf/s)
    gs: float = 0.0

    # Photosynthesis ------------------------------------------------------------

    # Maximum carboxylation rate (umol/m2/s)
    vcmax: float = 0.0

    # Maximum electron transport rate (umol/m2/s)
    jmax: float = 0.0

    # CO2 compensation point (umol/mol)
    cp: float = 0.0

    # Michaelis-Menten constant for CO2 (umol/mol)
    kc: float = 0.0

    # Michaelis-Menten constant for O2 (mmol/mol)
    ko: float = 0.0

    # Electron transport rate (umol/m2/s)
    je: float = 0.0

    # C4: Initial slope of CO2 response curve (mol/m2/s)
    kp_c4: float = 0.0

    # Leaf respiration rate (umol CO2/m2 leaf/s)
    rd: float = 0.0

    # Rubisco-limited gross photosynthesis (umol CO2/m2 leaf/s)
    ac: float = 0.0

    # RuBP regeneration-limited gross photosynthesis (umol CO2/m2 leaf/s)
    aj: float = 0.0

    # Product-limited (C3) or CO2-limited (C4) gross photosynthesis
    ap: float = 0.0

    # Leaf gross photosynthesis (umol CO2/m2 leaf/s)
    ag: float = 0.0

    # Leaf net photosynthesis (umol CO2/m2 leaf/s)
    an: float = 0.0

    # Leaf surface CO2 (umol/mol)
    cs: float = 0.0

    # Leaf intercellular CO2 (umol/mol)
    ci: float = 0.0

    # Leaf fractional humidity at surface (dimensionless)
    hs: float = 0.0

    # Leaf vapor pressure deficit at surface (Pa)
    vpd: float = 0.0

    # Model time step (s)
    dt: float = 0.0
```

## State variables objects (mutated every timestep)

SurEauPlant — split into State / Fluxes / Diagnostics
 
Design rationale (mirrors Bonan's Leaf → params, Flux → outputs):

1) SurEauPlantState: ODE state variables + quantities that persist across timesteps (the plant's "memory"). Initialised once, mutated by the solver.

2) SurEauPlantFluxes: Quantities computed fresh each (sub-)timestep from state + atmospheric forcing. Safe to zero-out between steps.

3) SurEauPlantDiagnostics: Solver quality metrics + derived outputs (fuel moisture) used for monitoring/logging.

The three objects replace the single SurEauPlant dataclass.

A thin SurEauPlant wrapper is provided at the bottom for backward-compatible attribute access during migration.
 

### State class: persists across timesteps 

```python
# | export


@dataclass
class SurEauPlantState:
    """Plant hydraulic state — mutated by the solver, carried forward in time.

    Contains the 4-compartment water potentials (leaf/stem × apoplasm/symplasm),
    cavitation fronts, conductances, capacitances, phenology, and tissue water
    content.  Everything here is needed to restart the simulation from a
    checkpoint.
    """

    # ── Water potentials [MPa] ───────────────────────────────────────────────
    psi_LApo: float = 0.0
    psi_SApo: float = 0.0
    psi_LSym: float = 0.0
    psi_SSym: float = 0.0

    # Cavitation fronts (most-negative ψ ever reached — irreversible)
    psi_LApo_cav: float = 0.0
    psi_SApo_cav: float = 0.0

    # Weighted soil boundary condition passed in from SurEauSoil
    psi_all_soil: float = 0.0

    # ── Hydraulic conductances ───────────────────────────────────────────────
    k_plant: float = 0.0
    k_LSym: float = 0.0
    k_SSym: float = 0.0
    k_RSApo: Optional[np.ndarray] = None  # per soil layer
    k_SLApo: float = 0.0
    k_soil_to_stem: Optional[np.ndarray] = None  # per soil layer

    # ── Capacitances ─────────────────────────────────────────────────────────
    C_LSym: float = 0.0
    C_SSym: float = 0.0
    C_LApo: float = 0.0
    C_SApo: float = 0.0

    # ── Cavitation (percent loss of conductivity) ────────────────────────────
    PLC_leaf: float = 0.0
    PLC_stem: float = 0.0

    # ── LAI & phenology ──────────────────────────────────────────────────────
    LAI_pheno: float = 0.0
    LAI: float = 0.0
    canopy_storage_capacity: float = 0.0
    FCC: float = 0.0
    LAI_dead: float = 0.0

    # ── Phenology accumulator ────────────────────────────────────────────────
    sum_temperature: float = 0.0
    budburst_date: float = np.nan

    # ── Tissue water content (saturated references & current) ────────────────
    #    *_sat_mmol / *_sat_L  : maximum storage at full turgor
    #    *_sat_mmol_per_LA     : same, normalised by leaf area
    #    Q_*_L                 : current water content [L m⁻² ground]

    Q_LApo_sat_mmol: float = 0.0
    Q_LApo_sat_L: float = 0.0
    Q_SApo_sat_mmol: float = 0.0
    Q_SApo_sat_L: float = 0.0
    Q_LSym_sat_mmol: float = 0.0
    Q_LSym_sat_L: float = 0.0
    Q_SSym_sat_mmol: float = 0.0
    Q_SSym_sat_L: float = 0.0
    Q_LApo_sat_mmol_per_LA: float = 0.0
    Q_SApo_sat_mmol_per_LA: float = 0.0
    Q_LSym_sat_mmol_per_LA: float = 0.0
    Q_SSym_sat_mmol_per_LA: float = 0.0
    Q_LApo_L: float = 0.0
    Q_SApo_L: float = 0.0
    Q_LSym_L: float = 0.0
    Q_SSym_L: float = 0.0

    # ── Canopy dry matter [g m⁻² ground] ────────────────────────────────────
    DM_live_canopy: float = 0.0
    DM_dead_canopy: float = 0.0
```

### Fluxes class: recomputed every (sub-)timestep 

```python
# | export


@dataclass
class SurEauPlantFluxes:
    """Instantaneous fluxes & conductances — computed from state + forcing.

    Everything here can safely be zeroed between timesteps; the solver
    fills them in from SurEauPlantState + atmospheric forcing.
    """

    # ── Stomatal / canopy conductance ────────────────────────────────────────
    gmin: float = 0.0  # cuticular conductance, T-adjusted [mmol m⁻² s⁻¹]
    gmin_S: float = 0.0  # stem cuticular conductance
    regul_fact: float = 0.01  # stomatal regulation factor [0–1]
    gs_bound: float = 0.0  # boundary-layer-limited stomatal cond.
    gs_lim: float = 0.0  # hydraulically limited stomatal cond.
    g_canopy_bound: float = 0.0
    g_canopy_lim: float = 0.0
    g_BL: float = 0.0  # boundary layer conductance
    g_crown: float = 0.0  # crown-level aerodynamic conductance

    # ── Water fluxes [mmol m⁻² s⁻¹ unless _mm suffix] ───────────────────────
    E_prime: float = 0.0  # atmospheric transpiration demand
    E_min: float = 0.0  # minimum (cuticular) leaf transpiration
    E_min_S: float = 0.0  # minimum stem transpiration
    E_bound: float = 0.0  # boundary-layer-limited transpiration
    E_lim: float = 0.0  # hydraulically limited transpiration

    flux_soil_to_stem: Optional[np.ndarray] = None  # per layer
    flux_soil_to_stem_mm: Optional[np.ndarray] = None  # per layer [mm]
    transpiration_mm: float = 0.0
    E_min_mm: float = 0.0
    E_min_S_mm: float = 0.0
    sum_flux_soil_to_stem: float = 0.0

    # ── Interception ─────────────────────────────────────────────────────────
    ppt_soil: float = 0.0  # throughfall reaching soil [mm]
    intercepted_water: float = 0.0  # water held in canopy [mm]
    evaporation_intercepted: float = 0.0  # evaporation of interception [mm]
    ETP_r: float = 0.0  # residual ETP after interception
    ETP: float = 0.0  # potential evapotranspiration

    # ── Leaf energy balance ──────────────────────────────────────────────────
    leaf_temperature: float = np.nan
    leaf_VPD: float = 0.0
```
