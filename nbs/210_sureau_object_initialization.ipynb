```python
# | export

def init_soil(soil_params):

    # Object initialization -----------------------------------------------------
    soil = SurEauSoil()

    # Prepare soil params -------------------------------------------------------
    soil.soil_water_stock = soil_params.V_field_capacity.copy()
    
    n = soil_params.n_layers

    soil.psi_soil = np.zeros(n) 
    soil.k_soil = np.zeros(n)
    soil.REW = np.zeros(n)
    
    return soil
```

```python
# | export

def init_plant(veg: SurEauVegetationParams):
    
    # Object initialization -----------------------------------------------------
    state = SurEauPlantState() 
    fluxes = SurEauPlantFluxes() 
    diag = SurEauPlantDiagnostics()
    
    # State: conductances -------------------------------------------------------
    state.k_plant = veg.k_plant_init 
    state.k_LSym = veg.k_LSym_init 
    state.k_SSym = veg.k_SSym_init
    
    # Fluxes: cuticular conductance default -------------------------------------
    fluxes.gmin_S = veg.gmin_S
    
    # Allocate per-layer arrays -------------------------------------------------
    n = len(veg.root_distribution) if veg.root_distribution is not None else 3
    
    state.k_soil_to_stem = np.zeros(n)
    fluxes.flux_soil_to_stem = np.zeros(n)
    fluxes.flux_soil_to_stem_mm = np.zeros(n)
    
    return state, fluxes, diag
 
```

```python
# | export

def yearly_init_plant(state: SurEauPlantState, 
                      fluxes: SurEauPlantFluxes, 
                      diag:SurEauPlantDiagnostics,
                      params: SurEauVegetationParams):

    # Water potentials ----------------------------------------------------------
    state.psi_LApo = 0.0 
    state.psi_SApo = 0.0 
    state.psi_LSym = 0.0 
    state.psi_SSym = 0.0
    
    state.psi_LApo_cav = 0.0
    state.psi_SApo_cav = 0.0 
    state.psi_all_soil = 0.0
    
    # Conductances --------------------------------------------------------------
    state.k_plant = params.k_plant_init 
    state.k_LSym = params.k_LSym_init 
    state.k_SSym = params.k_SSym_init
    
    # Cavitation ----------------------------------------------------------------
    state.PLC_leaf = 0.0 
    state.PLC_stem = 0.0
    
    # Phenology / LAI -----------------------------------------------------------
    # Evergreen species start each year at full canopy; deciduous/forced start 
    # bare
    if params.foliage == "Evergreen":
        state.LAI_pheno = params.LAI_max
    else:
        state.LAI_pheno = 0.0
        
    state.LAI = 0.0 
    state.LAI_dead = 0.0
    state.sum_temperature = 0.0 
    state.budburst_date = np.nan 
    state.DM_dead_canopy = 0.0
    
    # Reset fluxes --------------------------------------------------------------
    fluxes.gmin = 0.0
    fluxes.gmin_S = params.gmin_S
    fluxes.regul_fact = 0.01 
    fluxes.gs_lim = 0.0
    fluxes.E_prime = 0.0 
    fluxes.E_min = 0.0
    fluxes.E_min_S = 0.0
    fluxes.E_bound = 0.0 
    fluxes.E_lim = 0.0
    fluxes.transpiration_mm = 0.0
    fluxes.intercepted_water = 0.0
    fluxes.leaf_temperature = np.nan
    
    return state, fluxes, diag
```
