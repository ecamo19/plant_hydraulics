```python
# | export
from plant_hydraulics.leaf_stomatal_medlyn import leaf_stomatal_medlyn
from plant_hydraulics.parameter_classes import PhysCon, Leaf, Flux, Atmos
from plant_hydraulics.leaf_stomatal_profit_max import leaf_stomatal_profit_max
from plant_hydraulics.leaf_stomatal_optimization import leaf_stomatal_optimization
```

```python
# | export


def leaf_fluxes(physcon: PhysCon, atmos: Atmos, leaf: Leaf, flux: Flux) -> Flux:
    """
    Leaf fluxes — main entry point for computing leaf gas exchange.

    Dispatches to the appropriate stomatal model based on
    leaf.stomatal_model:

        "optimization" → Original WUE optimization (Bonan Ch. 13)
        "medlyn"       → Medlyn et al. (2011) empirical model
        "profit_max"   → Medlyn + Sperry profit-maximisation
    """
    if leaf.stomatal_model == "medlyn":
        return leaf_stomatal_medlyn(physcon, atmos, leaf, flux)
    elif leaf.stomatal_model == "profit_max":
        return leaf_stomatal_profit_max(physcon, atmos, leaf, flux)
    else:
        return leaf_stomatal_optimization(physcon, atmos, leaf, flux)
```
